```python
import scipy
import soundfile as sf
import numpy as np
```

```python
from transformers import AutoTokenizer, AutoModelForTextToWaveform
import torch

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("procit001/dutch_v1")
model = AutoModelForTextToWaveform.from_pretrained("procit001/dutch_v1")

def break_sentence_to_dict(sentence):
    sentences = sentence.split('.')
    result = {}
    index = 1
    for s in sentences:
        s = s.strip()
        if s:
            result[index] = s + ''
            index += 1
    return result

# Example usage
sentence = "Goedemorgen. Ik ben Annemarie, de digitale medewerker. Om a goed en snel van dienst te zijn, vragen wij om een aantal vragen te a beantwoorden. Hierdoor kunnen mijn collega's u beter helpen. Wilt u mijn vragen beantwoorden?"

result_dict = break_sentence_to_dict(sentence)

wave_list = []

for key, text in result_dict.items():
    # Tokenize the text
    inputs = tokenizer(text, return_tensors="pt")

    # Generate the waveform
    with torch.no_grad():
        output = model(**inputs).waveform

    # Convert the waveform to a 1D numpy array
    waveform = output.squeeze().cpu().numpy()
    wave_list.append(waveform)

    # Save the audio to a file
#     output_path = f"A_dutch_{key}.wav"
#     sf.write(output_path, waveform, model.config.sampling_rate)
# print("Audio files have been generated and saved.")
print(wave_list)
```

```python
from IPython.display import Audio
concatenated_audio = np.concatenate(wave_list)
audio=Audio(concatenated_audio, rate=model.config.sampling_rate*1.1)
audio
```
